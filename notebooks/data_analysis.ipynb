# Data Analysis Notebook

This notebook demonstrates data analysis using the sample dataset.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')

from scripts.preprocess import load_data, clean_data, get_statistics

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Load and Explore Data

```python
# Load data
df = load_data('../datasets/sample_data.csv')
print(df.head())
```

```python
# Data info
print(f"Shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nData types:\n{df.dtypes}")
```

## 3. Statistical Summary

```python
# Get statistics
df_clean = clean_data(df)
stats = get_statistics(df_clean)

print("Numeric Summary:")
print(df_clean.describe())
```

```python
print("\nMissing Values:")
print(df_clean.isnull().sum())
```

## 4. Sales Analysis by Product

```python
# Sales by product
sales_by_product = df_clean.groupby('product').agg({
    'sales': 'sum',
    'quantity': 'sum'
}).reset_index()

print(sales_by_product)
```

## 5. Regional Analysis

```python
# Sales by region
sales_by_region = df_clean.groupby('region').agg({
    'sales': ['sum', 'mean', 'count']
}).reset_index()

print(sales_by_region)
```

## 6. Customer Type Analysis

```python
# Sales by customer type
sales_by_type = df_clean.groupby('customer_type').agg({
    'sales': ['sum', 'mean', 'std']
}).reset_index()

print(sales_by_type)
```

## 7. Key Insights

```python
print("Key Metrics:")
print(f"Total Sales: ${df_clean['sales'].sum():,.2f}")
print(f"Average Sales per Transaction: ${df_clean['sales'].mean():,.2f}")
print(f"Total Quantity Sold: {df_clean['quantity'].sum()} units")
print(f"\nTop Product by Sales: {sales_by_product.loc[sales_by_product['sales'].idxmax(), 'product']}")
print(f"Top Region by Sales: {sales_by_region.loc[sales_by_region['sales'].idxmax(), 'region']}")
```
